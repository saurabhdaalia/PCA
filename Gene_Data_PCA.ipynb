### PCA Analysis for Gene Sample Data

### We know that Principal Component Analysis is used for dimension reduction for a data containing large set of features while retaining the most information.

#### PCA aims to transform the dataset by fiding its Eigen vector and Eigen values.

#### For this particular dataset we have considered 2 Principal components i.e PC1 and PC2. Principal Components helps us figure out variance in the dataset more efficienlty.  

### Total Variance for PC1 and PC2 comes out to be 33% which denotes that we might need more Principal Components to show more than 90% variance in the datasets.

###  We can clearly observe from the PC1 vs PC2 plot that different timepoints (at which different samples were taken) are differentiating. Samples taken at same timepoints shows similar Principal Components values thus similar variance.
